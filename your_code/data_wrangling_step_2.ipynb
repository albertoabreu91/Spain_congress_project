## Adding/changing to percentages tables
 
1- votes_per = t_votes / v_census * 100 into a column called abd_rate


2- Political party percentage = political party/t_votes * 100 


```python
# Getting the total voting percentage

spain_pol['votes_rate'] = (spain_pol['t_votes'] / spain_pol['v_census']) * 100
```

```python
spain_pol.columns
```

```python
for columns in spain_pol:
    print(spain_pol[columns])
```

```python
# Create a parties table with all the political parties column names
# Creata a total votes list with all the total votes per municips


parties = list(spain_pol.columns)[8:-2]
t_votes_list = list(spain_pol["t_votes"])

# For loop to iterate through the DataFrame to create new columns with the voting percentage per party.

for i in range(len(parties)):
    name = parties[i]+"_per"
    spain_pol[name] = spain_pol[parties[i]]/spain_pol["t_votes"] * 100
```
