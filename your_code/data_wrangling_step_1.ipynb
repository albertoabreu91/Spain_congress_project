```python
import pandas as pd
import numpy as np
```

```python
# Importin Origin table spanish_regions.csv

spanish_regions = pd.read_csv('../spanishregions_dataset.csv', sep=';')
```

```python
# Changing the name of "La Vall de Boí" to "Vall de Boí " to have the same name as in the muni_table shown next"

spanish_regions.loc[3847, 'RegionName'] = 'Vall de Boí'
```

```python
spanish_regions.head()
```

```python
spanish_regions[spanish_regions['RegionName'] == 'Vall de Boí']
```

## Importing the muni_table data containing information about municipalities in Catalonia

```python
muni_table = pd.read_csv('../muni_table.csv')
```

```python
muni_table.head()
```

```python
# Changing the name of the Municipality column

muni_table = muni_table.rename(columns= {'Municipality': 'RegionName'})
muni_table.info()
```

```python
muni_table = muni_table[['Province', 'RegionName']][:-1]
```

```python
muni_table.sort_values(['Province', 'RegionName']).head()
```

#### Joining the two tables spanish_regions and muni_table to see if they have repeated RegionNames

```python
join_tables = spanish_regions.merge(muni_table, how= 'right')
```

```python
join_tables.head()
```

```python
join_tables[join_tables['RegionName'] == 'Vall de Boí']
```

```python
join_tables[join_tables['RegionName'].duplicated()]
```

```python
join_tables[join_tables['RegionName'] == 'Cabanes']
```
